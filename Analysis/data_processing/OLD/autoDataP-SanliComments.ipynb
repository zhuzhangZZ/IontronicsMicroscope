#SF: suggested analysis procedure
For each salt/region
* 1- have a first glance at the variance and choose 3-4 interesting spots (usually localized) based on square-wave movie
* 2- if there is a region that has homogeneous by high var I/I, also include that in one analysis
* 3- for all the movies in the series, run all the cells for each of the spots and also save all these images, group images for clarity
    * group 1, overview
        * average of center line vs time 
        * average image full area
        * log10(varI/I) image full area
    * group 2x, x for each selected region of roughly 150 x 150 pix
        * average image selected area
        * intensity of max point vs time
        * fft components of I vs t for spot_max
    * group 3x, x for each selected region of roughly 150 x 150 pix   
        * first harmonic for selected area 
        * oscillation phase for selected area   
        * mean intensity for selected area (same as group 2x, here for comparison)
        * ratio of first harmonic to mean intensity for selected area 
    * group 4x, for each bright spot 
        * intensity of the whole over the spot vs time (certainly save data and npy arrays)
* 4- For the each selected spot, use the extracted data to compare the intensity vs time for all the conditions in the group of movies from that region, e.g. DI vs salt, vs different waveform. For each sequence normalize the curves to the average intensity (but print the average in the graph legend)

## Finding the harmonic frequency and looking at the spectrum:
#SF: this indeed might take a bit of time, but not so much of chosen SOI is smaller than 200 x 200

#SF: do not put the FFT for each pixel in the loop, that's inefficient. use the numpy function for FFT over the whole movie

```python
maxFFTc = 100 # fourier index above which is irrelevant for the analysis
dcbase = 5   # fourier index below which counts as drift
safestartframe, safeendframe = 0, 2400
    
partOfMovieForFFT = np.array(movie[:,:,safestartframe:safeendframe]).copy()
FFTspecSOI = np.fft.rfft(partOfMovieForFFT[xminf:xmaxf,yminf:ymaxf,:], axis=2)/(safeendframe - safestartframe)
# Find maximum valued frequency
mfreq = dcbase + np.argmax(abs(FFTspecSOI[max_var_pos_x, max_var_pos_y, dcbase:maxFFTc]))
#shfreq = mfreq + dcbase + np.argmax(abs(fspec[mfreq + dcbase:]))
#thirdfreq = shfreq + dcbase + np.argmax(abs(fspec[shfreq + dcbase:]))
print("Fourier components:", mfreq)


plt.plot(np.arange(dcbase, maxFFTc), np.log10(abs(FFTspecSOI[max_var_pos_x, max_var_pos_y, dcbase:maxFFTc]))) # Give correct "x"axis
plt.title("Fourier components of interest")
plt.xlabel("Fourier component")
plt.ylabel("Amplitude?")
plt.show()
```

## Another plot:
Another way to summerize the data is Sum\[I(t).sin(wt)\] where w is the 
modulation frequency in 1/frames and the sum is taken over all images.

```python
def getSumIsinW(partOfMovieForFFTVarPlot, mfreq):
    sumitsinwt = np.zeros_like(partOfMovieForFFTVarPlot[:,:,0])
    for i in range(len(partOfMovieForFFTVarPlot[:,0,0])):
        for j in range(len(partOfMovieForFFTVarPlot[0,:,0])):
            sumitsinwt[i,j] = sum([partOfMovieForFFTVarPlot[i,j,t]*np.sin(1/mfreq*t) for t in range(len(partOfMovieForFFTVarPlot[0,0,:]))])
    return sumitsinwt

sumitsinwt = getSumIsinW(partOfMovieForFFTVarPlot, mfreq)
plt.imshow(sumitsinwt, extent=(yminf,ymaxf,xminf,xmaxf), origin='lower')
plt.title("Sum[I(t).sin(wt)] dt")
plt.xlabel("x")
plt.ylabel("y")
if save_automatically: 
    figurename = "Sum-It-sinwt-dt"
    customSave(figurename, sumitsinwt)
plt.show()
```
